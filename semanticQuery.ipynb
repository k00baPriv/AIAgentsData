```python
from datamart import SemanticQuery, run_semantic_query
from datamart.db import get_connection

connection = get_connection()
rows = run_semantic_query(
    connection,
    SemanticQuery(metric="payments_collected", dimensions=("payment_method",), day_offset=1),
)
print(rows)
```

```python
rows = run_semantic_query(
    connection,
    SemanticQuery(
        metric="net_revenue",
        dimensions=("category",),
        aggregate_over_time=True,
    ),
)

for row in rows:
    print(row)
```

```python
rows = run_semantic_query(
    connection,
    SemanticQuery(
        metric="new_customers",
        grain="day",
        trailing_days=21,
    ),
)

for row in rows:
    print(row)
```

```python
rows = run_semantic_query(
    connection,
    SemanticQuery(
        metric="new_customers",
        grain="week",
        trailing_days=21,
    ),
)

for row in rows:
    print(row)
```

```python
rows = run_semantic_query(
    connection,
    SemanticQuery( metric="net_revenue", grain="week", compare_to="previous_period", ),
)
```

```python
for row in rows:
    print(row)
```
